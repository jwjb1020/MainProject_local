glob로 폴더에 있는 이미지 파일 리스트로 묶은 뒤 plt를 써서 각각 보여주기

```python
from PIL import Image
import matplotlib.pyplot as plt
import glob as gl

image_list = []
for filename in gl.glob('runs/detect/predict/*.jpg'):
    im = Image.open(filename)
    image_list.append(im)

# 이미지를 순회하며 각각을 표시
for im in image_list:
    plt.figure()
    plt.imshow(im)
    plt.show()
```
